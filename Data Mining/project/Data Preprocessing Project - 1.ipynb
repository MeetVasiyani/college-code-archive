#### 11. Remove Missing Customer IDs

```python
df = df.dropna(subset=['CustomerID'])
df
```

#### 12. Remove Duplicates from the Dataset

```python
df.drop_duplicates(inplace=True)
df
```

#### 13. Change Datatypes of few Columns

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

#### 14. Add Total Price Column

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df
```

#### 15. Remove Negative or Zero Prices/Quantities

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df
```

#### 16. Most Frequent Customers

```python
df['CustomerID'].value_counts().head(10)
```

#### 17. Average Items per Invoice

```python
avg_items = df.groupby('InvoiceNo')['Quantity'].sum().mean()
print("Average Number of items : ",avg_items)
```

#### 18. Correlation Matrix

```python
df[['Quantity', 'UnitPrice', 'TotalPrice']].corr()
```

#### 19. Covariance Matrix

#### 20. Most Sold Product

```python
most_sold_product = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
most_sold_product
```

#### 21. Highest Total Price Order

```python
df.sort_values(by='TotalPrice',ascending=False)
```

#### 22. Same Top N Products with Highest Unit Price

#### 23. Plotting Most Sold Products

#### 24. Most Revenued Countries

```python
most_revenue_country = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
most_revenue_country.plot(kind='bar')
```
